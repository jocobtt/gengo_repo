```python
import pymc4 as pm4
import pandas as pd 
import numpy as np 
import seaborn as sns 
import os 
import matplotlib.pyplot as plt
import arviz as az 
```

```python
myu = 8
sig = 2.5

data = np.random.normal(loc=myu, scale=sig, size=200)
```

```python
# model def 
@pm4.model
def model(data):
    myu = yield pm4.Normal(loc=0, scale=10, name='mu')
    sig = yield pm4.Exponential(rate=0.1, name='sig')
    
    like = yield pm4.Normal(loc=myu, scale=sig, observed=data, name='like')
    
    return like
```

```python
# sample from posterior 
estimation_model = model(data)

trace = pm.sample(model(data), num_samples=800)
```

```python
az.plot_posterior(trace, var_names='model/mu');
```

```python
az.plot_posterior(trace, var_names='model/__log_sig');
```

```python
# prior/posterior predictive samples
draws_prior = pm4.sample_prior_predictive(estimation_model)
```

```python
ax = az.plot_kde(draws_prior.prior_predictive['model/like'], label='prior_predictive');
ax = az.plot_kde(data, label='data', plot_kwargs={'color':'C1'});
```

```python
draws_prior
```

```python
draws_prior.prior_predictive
```

```python
draws_posterior = pm4.sample_posterior_predictive(estimation_model, trace,
                                                 inplace=False)
```

```python
draws_posterior
```

```python
draws_posterior.posterior_predictive
```

```python
combined_trace = trace + draws_posterior + draws_prior
```

```python
az.plot_ppc(combined_trace);
```

```python
# arbitrary curve regression on decay dataset 

# Constants for load_decay
tau = 71.9 # indium decay half life
A = 42  # starting magnitude
C = 21 # measurement error
noise_scale = 1


def load_decay():
    t = np.arange(0, 1000)
    def decay_func(ts, noise):
        return A  * np.exp(-t/tau) + C + np.random.normal(0, noise, size=(len(t)))

    data = {'t': t, 'activity': decay_func(t, noise_scale)}
    df = pd.DataFrame(data)
    return df


```

```python
np.random.seed(303)

df = load_decay()
df.head(5)
```

```python
ax = df['activity'].plot()
```
